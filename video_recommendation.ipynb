```python
import numpy as np
import pandas as pd
```

```python
metadata = pd.read_csv('VIDEOs.csv')
metadata.head(9)
```

```python
C = metadata['views'].mean()
print(C)
```

```python
m = metadata['likes'].quantile(0.90)
print(m)
```

```python
q_videos = metadata.copy().loc[metadata['views'] >= m]
q_videos.shape
```

```python
metadata.shape
```

```python
def weighted_rating(x, m=m, C=C):
    v = x['views']
    R = x['likes']
    return (v/(v+m) * R) + (m/(m+v) * C)
```

```python
q_videos['score'] = q_videos.apply(weighted_rating, axis=1)
```

```python
q_videos = q_videos.sort_values('score', ascending=False)
q_videos[['title', 'likes', 'views', 'score']].head(5)
```
